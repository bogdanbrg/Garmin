```python
# Cell: Create database
import sqlite3

db_path = r'C:\Users\Svitlana\OneDrive\Garmin\garmin.db'

conn = sqlite3.connect(db_path)
conn.close()

print("✅ Database created")
```

```python
# Cell: Create empty table

conn = sqlite3.connect(db_path)
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE raw_get_gear (
    uuid TEXT,
    customMakeModel TEXT, 
    gearTypeName TEXT,
    gearStatusName TEXT
)
""")

conn.commit()
conn.close()

print("✅ Table 'raw_get_gear' created")
```

```python
# Cell: Delete table

conn = sqlite3.connect(db_path)
cursor = conn.cursor()

cursor.execute("DROP TABLE raw_get_gear")

conn.commit()
conn.close()

print("✅ Table 'raw_get_gear' deleted")
```
